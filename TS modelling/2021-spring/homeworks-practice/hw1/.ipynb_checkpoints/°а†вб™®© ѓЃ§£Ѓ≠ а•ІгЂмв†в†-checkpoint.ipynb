```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as  np
import seaborn as sns
```

```python
futers = pd.read_csv('C:\\Users\\admin\\anaconda3\\пфи\\сем1\\futers.csv', sep=',')
futers.columns=['Ticker','PER','DATE','TIME','OPEN','HIGH','LOW','CLOSE','VOL']
spot=pd.read_csv('C:\\Users\\admin\\anaconda3\\пфи\\сем1\\spot.csv', sep=',')
spot.columns=['Ticker','PER','DATE','TIME','OPEN','HIGH','LOW','CLOSE','VOL']
```

```python
futers.DATE=pd.to_datetime(futers.DATE, format='%d/%m/%y')
spot.DATE=pd.to_datetime(spot.DATE, format='%d/%m/%y')
```

```python
futers.set_index('DATE', inplace=True)
spot.set_index('DATE', inplace=True)
```

```python
spot
```

```python
# обрезал первые 3 наблюдения потому что их базис больше нуля и они пиздец в картину не вписываются

futers=futers.iloc[3:,6]
spot=spot.iloc[3:,6]
```

```python
basis=spot-futers
```

```python
basis=basis.interpolate(method='linear')
```

```python
basis
```

```python
from sklearn.linear_model import LinearRegression
```

```python
model=LinearRegression()
```

```python
model.fit(X=np.arange(len(basis)).reshape(-1,1), y=basis.values)
```

```python
# выбрал 30 потому что иначе последнее наблюдение получится меньше нуля и у меня не получится пересекающаяся с нулем линия

y_pred=model.predict(X=np.arange(len(basis),len(basis)+30).reshape(-1,1))
```

```python
y_pred
```

```python
fig, ax=plt.subplots(figsize=(15,6))

sns.regplot(x=np.arange(len(basis)), y=basis.values, label='historic basis')
sns.scatterplot(x=np.arange(len(basis),len(basis)+30), y=y_pred, color='y', label='predicted basis')
ax.hlines(0, xmin=0, xmax=(len(basis)+36))
ax.legend(loc='lower right')
```
